### Filter the dataset, remove unused, inconsistent and outliers values.

Filter only residential register type and drop unecessary columns

```python
# filter only residential
df = df[df['property_type'] == 'residential']

# drop unnecessary columns, these columns won't be useful in analysis and prediction
df = df.drop(['apn','property_type'], axis=1)
```

Remove registers sqft and price with zero values. Remove zip_code empty too.

```python
# Not necessery remove 'lot', because this value will not be used. Low correlation
#df = df[df['lot'] > 0.1]

# remove sqft zero
df = df[df['sqft'] > 0.1]

# remove price zero
df = df[df['price'] > 0.1]

# Remove residential with zipcode null
df = df[df['zip_code'].notnull()]

# Converto all to float
df['beds'] = df['beds'].astype(float)
df['price'] = df['price'].astype(float)
```

Remove outliers values, values with greater than 3x standard deviation. This process is useful for removing values out of the analysis, avoiding negative effects. It may be typing error or even unusual values

```python
for var in ['baths', 'beds', 'sqft', 'price']:
    df = df[df[var] < df[var].mean()+3*df[var].std()]
```

#### Converts postal code to coordinate latitude and longitude

```python
# Zip codes
df_zip = pd.read_csv("zipcod_to_cord.txt")

def zipcode_to_lat_lng(zipcode):
    zipnum = int(zipcode.split("-")[0])
    line = df_zip[df_zip['ZIP'] == zipnum]
    if len(line) > 0:
        return [line.LAT.values[0], line.LNG.values[0]]
    else:
        return [None, None]
```

This process may take a few minutes because it uses a webservice, but only runs once. Do not need to use it if it were in production, data would already be available

```python
df['lat'] = df['zip_code'].apply(lambda x: zipcode_to_lat_lng(x)[0])
df['lng'] = df['zip_code'].apply(lambda x: zipcode_to_lat_lng(x)[1])
```

```python
# Remove residential with lat or lng is null
df = df[df['lng'].notnull()]
```

### Choose 4 models from different categories (Machine Learning)

I chose a regions model (RandomForestRegressor and KNN), linear with LinearRegression and nolinear with SVN and Kernel poly. The problem is a non-linear problem, Mainly by the variables latitude and longitude. 

#### Machine Learning 

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn import neighbors
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
from sklearn import preprocessing
```

```python
# funcoes
def plot_residuo(residuo):
    plt.scatter([x for x in range(0,len(residuo))], residuo)    
```

```python
# Scores model
score = {}
```

#### Split  dataset in train and test

Split dataset to TRAIN (70%) and TEST (30%) data

```python
# Split 30% for test
X_train, X_test, y_train, y_test = train_test_split(df[['lat', 'lng', 'sqft', 'beds', 'baths']]
                                                    ,df.price, test_size=0.30, random_state=42)
```

```python
len(X_train)
```

```python
len(X_test)
```

#### Normalized

Normalize dataset with StandardScaler. Is necessary for the proper functioning of some algorithms

```python
scaler = preprocessing.StandardScaler().fit(X_train)

t_X_train = scaler.transform(X_train)  
t_X_test = scaler.transform(X_test)
```

#### Linear Regression

```python
# Create linear regression object
model1 = linear_model.LinearRegression(normalize=False)
```

```python
# Train the model using the training sets
model1.fit(t_X_train, y_train)
```

```python
# The mean squared error
print("Mean squared error: %.2f" % np.mean((model1.predict(t_X_test) - y_test) ** 2))

# Explained variance score: 1 is perfect prediction
score['linear'] = model1.score(t_X_test, y_test)
print('Variance score: %.2f' % score['linear'])
```

```python
# Plot residuo 
plot_residuo(model1.predict(t_X_test) - y_test)
```

#### KNN

```python
# Create nolinear regression object with knn
knn = neighbors.KNeighborsRegressor(10, weights='distance', metric='euclidean')

knn.fit(t_X_train, y_train)
```

```python
# The mean squared error
print("Mean squared error: %.2f" % np.mean((knn.predict(t_X_test) - y_test) ** 2))

# Explained variance score: 1 is perfect prediction
score['knn'] = knn.score(t_X_test, y_test)
print('Variance score: %.2f' % score['knn'])
```

```python
plot_residuo(knn.predict(t_X_test) - y_test)
```

#### Random Forest Regressor

```python
forest = RandomForestRegressor(n_estimators=100)

forest.fit(t_X_train, y_train)
```

```python
# The mean squared error
print("Mean squared error: %.2f" % np.mean((forest.predict(t_X_test) - y_test) ** 2))

# Explained variance score: 1 is perfect prediction
score['RandomForest'] =  forest.score(t_X_test, y_test)
print('Variance score: %.2f' % score['RandomForest'])
```

```python
plot_residuo(forest.predict(t_X_test) - y_test)
```

#### SVN

```python
svn = SVR(kernel='poly')

svn.fit(t_X_train, y_train)
```
